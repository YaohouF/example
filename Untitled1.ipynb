```python
import numpy as np 
np.seterr(divide='ignore', invalid='ignore')
import cv2
import matplotlib.pyplot as plt
import os
import time
import math

rootpath='./'
left_img = cv2.imread(os.path.join(rootpath, "left_image.jpg"), cv2.IMREAD_GRAYSCALE)
right_img = cv2.imread(os.path.join(rootpath, "right_image.jpg"), cv2.IMREAD_GRAYSCALE)
dis_image = cv2.imread(os.path.join(rootpath, "disparity.png"), cv2.IMREAD_ANYDEPTH)
ground_truth = dis_image/256
```

```python
# cost functions

def ssd(left_block, right_block):
    if left_block.shape != right_block.shape:
        return -1

    return np.sum((left_block - right_block)**2)

def ncc(left_block, right_block, left_bar, right_bar):
    if left_block.shape != right_block.shape:
        return -1
    l_r = 0
    l_var = 0
    r_var = 0
    left_block = left_block.flatten()
    right_block = right_block.flatten()
    l = left_block - left_bar
    r = right_block - right_bar
    l_r = l*r
    l_var = l*l
    r_var = r*r

   # for i in range(h):
       # for j in range(w):
          #  l = left_block[i,j] - left_bar
          #  r = right_block[i,j] - right_bar
          #  l_r += l*r
           # l_var += l*l
           # r_var += r*r
    d = np.sqrt(np.sum(l_var*r_var))
    if d != 0:
        result = np.sum(l_r)/d
    else:
        result = 0.0
    return result 
```

```python
# analyses functions

def rmse(disparity_map, ground_truth):

    idx = np.nonzero(ground_truth)
    valid_ground_truth = ground_truth[idx]
    valid_disparity_map = disparity_map[idx]

    rmse = np.linalg.norm(valid_disparity_map - valid_ground_truth) / np.sqrt(len(valid_ground_truth))

    return rmse


def error_fraction(disparity_map, ground_truth, threshold):
    
    idx = np.nonzero(ground_truth)
    valid_ground_truth = ground_truth[idx]
    valid_disparity_map = disparity_map[idx]

    fraction = np.sum(abs(valid_ground_truth - valid_disparity_map) < threshold)/len(valid_ground_truth)

    return fraction

threshold = [4, 2, 1, 0.5, 0.25]
```

```python
# without pre or post processing, search within a defined window

def stereo_maching(left_img, right_img, block_size, window, cost_func):
    """
    params: left image, right image,
            cost function: choose between SAD and SSD
            block size: dimension of the square kernel, odd number
            window: range of search
    return: disparity map between the left and the right image
            runtime per image
    """

    start = time.time()

    (hl, wl) = left_img.shape
    (hr, wr) = right_img.shape
    half_size = int(block_size/2)

    if (hl, wl) != (hr, wr):
        return -1
    
    else:
        disparity_map = np.zeros((hl, wl), dtype = np.uint16)

        """
        Question: why not starting from 0?
        """
        for y in range(half_size, hl-block_size):
            """
            Question: why not starting from 0?
            """
            if cost_func == 'NCC':
                left_bar = 0
                right_bar = 0
                n = 0
            for x in range(half_size, wl-block_size):
                left_block = left_img[max(y - half_size, 0) : min(y + half_size + 1, hl), max(x - half_size, 0) : min(x + half_size + 1, wl)]
                left_bar = np.sum(left_block)/left_block.size
                #min_diff = float('inf')
                min_diff = float('-inf') #uint16 size
                index = 0

                for rx in range(max(0, x - window), min(wr, x + window)):
                    right_block = right_img[max(y - half_size, 0) : min(y + half_size + 1, hl), max(rx - half_size, 0) : min(rx + half_size + 1, wl)]
                    right_bar = np.sum(right_block)/right_block.size

                    #diff = ssd(left_block, right_block) if cost_func == 'SSD' else sad(left_block, right_block)
                    diff = ncc(left_block, right_block, left_bar, right_bar)
                    #if math.isnan(diff):
                    #if diff < min_diff:
                       # min_diff = diff
                       # index = rx
                    if diff > min_diff:
                        min_diff = diff
                        index = rx
                disparity_map[y, x] = abs(index - x)

    end = time.time()
    runtime = end - start
        
    return disparity_map, runtime
```

```python
ncc_map, runtime_ncc = stereo_maching(left_img, right_img, 15, 32, 'NCC')
```

```python
rmse_ncc = rmse(ncc_map, ground_truth)
print(f'runtime = {runtime_ncc}')
print(f'RMSE = {rmse_ncc}')
```

```python
for i in threshold:
    error_frac = error_fraction(ncc_map, ground_truth, i)
    print(f'The fraction of pixels with error less than {i} is: {error_frac}')
```
